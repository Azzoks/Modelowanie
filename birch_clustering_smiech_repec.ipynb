```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import Birch
```

```python
data = load_iris(as_frame = True).frame

data
```

```python
model = Birch(branching_factor = 50, n_clusters = None, threshold = 1.5)
```

```python
model.fit(data)
```

```python
pred = model.predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import Birch
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data
y = iris.target


birch = Birch(n_clusters=3)  # n_clusters=3, poniewaÅ¼ wiemy, Å¼e sÄ… 3 klasy w Iris
birch.fit(X)
labels = birch.labels_
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Wizualizacja klastrÃ³w
plt.figure(figsize=(8, 6))
for cluster in np.unique(labels):
    plt.scatter(X_pca[labels == cluster, 0], X_pca[labels == cluster, 1], label=f'Cluster {cluster}')

plt.title("BIRCH Clustering of Iris Dataset")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend()
plt.show()
```
